Cluster ordering function adapted from Ishimoto 2018 Sci Rep

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
from scipy.optimize import curve_fit
```

```python
# --- Configuration ---
BOX_SIZE = 500.0  # Must match the box size of the simulation
R_MAX = BOX_SIZE/32     # Max distance to calculate correlation for
DR = R_MAX/200          # Bin width for distance (resolution)
```

```python
def compute_rdf_pbc(coords1, coords2, box_size, r_range):
    """
    Computes the 2D Radial Distribution Function g(r) with Periodic Boundary Conditions.
    
    Args:
        coords1: (N, 2) array of positions for reference particles (Type A)
        coords2: (M, 2) array of positions for target particles (Type B)
        box_size: Side length of the simulation box
        r_range: Array of bin edges for distance
    """
    
    # 1. Calculate vector differences (N x M x 2) using broadcasting
    delta = coords1[:, np.newaxis, :] - coords2[np.newaxis, :, :]
    
    # 2. Apply Periodic Boundary Conditions (Minimum Image Convention)
    delta = delta - box_size * np.round(delta / box_size)
    
    # 3. Calculate squared distances
    dists_sq = np.sum(delta**2, axis=2)
    dists = np.sqrt(dists_sq)
    
    # 4. Flatten
    dists_flat = dists.flatten()
    dists_flat = dists_flat[dists_flat > 1e-6] # Remove self-interactions
    
    # 5. Histogram counts
    counts, bin_edges = np.histogram(dists_flat, bins=r_range)
    
    # 6. Normalization by Ideal Gas Density (Area of shells)
    # Area of 2D shell = pi * (r_outer^2 - r_inner^2)
    areas = np.pi * (r_range[1:]**2 - r_range[:-1]**2)
    
    # Global density of Type B
    rho_2 = len(coords2) / (box_size**2)
    
    # Expected count if particles were randomly distributed
    # Density_target * Area_shell
    expected_counts = rho_2 * areas
    
    # Calculate g(r)
    g_r = counts / (len(coords1) * expected_counts)       
    
    return g_r

def analyze_and_plot(filename, times_to_analyze, cmap, heterogenous=False):
    """Loads a parquet file, computes correlations, and plots them."""
    print(f"--- Analyzing: {filename} ---")
    df = pd.read_parquet(filename)
    
    # Setup bins
    r_bins = np.arange(0, R_MAX + DR, DR)

    # Get bin centers for plotting
    r = (r_bins[:-1] + r_bins[1:]) / 2

    # Prepare Plot
    fig, ax = plt.subplots(3 if heterogenous else 1, 1, figsize=(6, 5*(3 if heterogenous else 1)), sharex=True, sharey=True)
    if not heterogenous:
        ax = [ax]
    
    for i, t in enumerate(times_to_analyze):
        df_t = df[df['Time'] == t]

        if not heterogenous:
            # Extract positions
            pos_all = df_t[['Pos_X', 'Pos_Y']].values
    
            # --- A. All-to-All Correlation ---
            # Useful for homogeneous simulations
            if len(pos_all) > 0:
                g = compute_rdf_pbc(pos_all, pos_all, BOX_SIZE, r_bins)
                ax[0].plot(r, g, label=f't={t}', color=cmap((i+1)/len(times_to_analyze)))
                ax[0].set_title(f"File: {filename.split(".")[0]}")

        if heterogenous:
            # Extract positions
            pos_motile = df_t[df_t['Type'] == 'motile'][['Pos_X', 'Pos_Y']].values
            pos_sub = df_t[df_t['Type'] == 'submotile'][['Pos_X', 'Pos_Y']].values
            
            # --- B. Specific Correlations (for Heterogeneous) ---
            # 1. Motile-Motile
            if len(pos_motile) > 10:
                g = compute_rdf_pbc(pos_motile, pos_motile, BOX_SIZE, r_bins)
                ax[0].plot(r, g, label=f't={t}', color=cmap((i+1)/len(times_to_analyze)))
                ax[0].set_title('Motile-Motile')
                
            # 2. Submotile-Submotile
            if len(pos_sub) > 10:
                g = compute_rdf_pbc(pos_sub, pos_sub, BOX_SIZE, r_bins)
                ax[1].plot(r, g, label=f't={t}', color=cmap((i+1)/len(times_to_analyze)))
                ax[1].set_title('Submotile-Submotile')
                
            # 3. Motile-Submotile (Cross Correlation)
            if len(pos_motile) > 0 and len(pos_sub) > 0:
                g = compute_rdf_pbc(pos_motile, pos_sub, BOX_SIZE, r_bins)
                ax[2].plot(r, g, label=f't={t}', color=cmap((i+1)/len(times_to_analyze)))
                ax[2].set_title('Motile-Submotile')

            plt.suptitle(f"File: {filename.split(".")[0]}")

    for i in ax:
        i.axhline(1.0, color='gray', linestyle=':', alpha=0.5) # Reference line
        i.legend()
        i.set_xlabel("Distance $r$")
        i.set_ylabel("$C_*(r)$")

    plt.tight_layout()

def analyse_all_time(filename, heterogenous=False):
    print(f"--- Analyzing: {filename} ---")
    df = pd.read_parquet(filename)
    times_to_analyze = df['Time'].unique()
    
    # Setup bins
    r_bins = np.arange(0, R_MAX + DR, DR)
    g_list = []
    
    for t in times_to_analyze:
        df_t = df[df['Time'] == t]

        if not heterogenous:
            # Extract positions
            pos_all = df_t[['Pos_X', 'Pos_Y']].values
    
            # --- A. All-to-All Correlation ---
            # Useful for homogeneous simulations
            if len(pos_all) > 0:
                g_list.append({
                    "Time": t,
                    "Cluster ordering": compute_rdf_pbc(pos_all, pos_all, BOX_SIZE, r_bins)
                })

        if heterogenous:
            # Extract positions
            pos_motile = df_t[df_t['Type'] == 'motile'][['Pos_X', 'Pos_Y']].values
            pos_sub = df_t[df_t['Type'] == 'submotile'][['Pos_X', 'Pos_Y']].values
            
            # --- B. Specific Correlations (for Heterogeneous) ---
            # 1. Motile-Motile
            if len(pos_motile) > 10:
                g_mm = compute_rdf_pbc(pos_motile, pos_motile, BOX_SIZE, r_bins)
                
            # 2. Submotile-Submotile
            if len(pos_sub) > 10:
                g_ss = compute_rdf_pbc(pos_sub, pos_sub, BOX_SIZE, r_bins)
                
            # 3. Motile-Submotile (Cross Correlation)
            if len(pos_motile) > 0 and len(pos_sub) > 0:
                g_ms = compute_rdf_pbc(pos_motile, pos_sub, BOX_SIZE, r_bins)

            g_list.append({
                "Time": t,
                "Cluster ordering: Motile-Motile": g_mm,
                "Cluster ordering: Submotile-Submotile": g_ss,
                "Cluster_ordering: Motile-Submotile": g_ms
            })

    return g_list
```

```python
# --- 1. Define the Log-Normal Model ---
def log_normal_shifted(r, A, mu, sigma, C):
    """
    Model: g(r) = C + (A / r) * exp(- (ln(r) - mu)^2 / (2 * sigma^2))
    
    Parameters:
    A     : Area under the peak (Strength of correlation)
    mu    : Log-scale mean (related to peak position)
    sigma : Shape parameter (width/skewness of the peak)
    C     : Baseline (should settle to 1.0)
    """
    # Prevent divide by zero or log of zero/negative
    r_safe = np.maximum(r, 1e-6)
    
    coeff = A / (r_safe * sigma * np.sqrt(2 * np.pi))
    exponent = -((np.log(r_safe) - mu)**2) / (2 * sigma**2)
    
    return C + coeff * np.exp(exponent)

# --- 2. Fitting Function ---
def fit_rdf_lognormal(r_data, g_data, fit_range=(0.0, 50.0)):
    """
    Fits the RDF to a Log-Normal peak within a specific range.
    """
    # 1. Slice data to the relevant range (usually just the first big peak)
    mask = (r_data >= fit_range[0]) & (r_data <= fit_range[1])
    r_fit = r_data[mask]
    g_fit = g_data[mask]
    
    if len(r_fit) < 5:
        return None, None
    
    # 2. Intelligent Initial Guesses
    # Peak location (approximate mu)
    peak_r = r_fit[np.argmax(g_fit)]
    mu_guess = np.log(peak_r)
    
    # Amplitude guess
    A_guess = np.max(g_fit) * 10.0 
    
    # p0 = [A, mu, sigma, C]
    p0 = [A_guess, mu_guess, 0.5, 1.0]
    
    # Bounds: A>0, sigma>0, C approx 1
    bounds = ([0, -np.inf, 1e-6, 0.8], [np.inf, np.inf, 5.0, 1.2])

    try:
        popt, pcov = curve_fit(log_normal_shifted, r_fit, g_fit, p0=p0, bounds=bounds, maxfev=5000)
        return popt, r_fit
    except Exception as e:
        print(f"Fit failed: {e}")
        return None, None

# --- 3. Analysis Wrapper ---
def analyze_peak_shape(filename, time_point):
    print(f"--- Peak Analysis for {filename} ---")
    df = pd.read_parquet(filename)
    
    # Get Time
    times = df['Time'].unique()
    t_actual = times[np.argmin(np.abs(times - time_point))]
    df_t = df[df['Time'] == t_actual]
    
    # Setup Bins (Fine resolution for better fitting)
    dr = 0.5 
    r_bins = np.arange(0, 80.0, dr)
    
    # Extract Positions (e.g., Motile-Motile)
    pos = df_t[df_t['Type'] == 'motile'][['Pos_X', 'Pos_Y']].values
    
    if len(pos) > 10:
        # Compute RDF
        r, g = compute_rdf_pbc(pos, pos, 500.0, r_bins, exclude_self=True)
        
        # --- FIT THE PEAK ---
        # Adjust fit_range to cover your main hump. 
        # range=(5, 60) usually covers the rise and the fall of the first shell.
        popt, r_fit_domain = fit_rdf_lognormal(r, g, fit_range=(5.0, 60.0))
        
        # Plot
        plt.figure(figsize=(8, 5))
        plt.plot(r, g, 'k.', alpha=0.4, label='Data')
        
        if popt is not None:
            A, mu, sigma, C = popt
            
            # Generate smooth curve
            r_smooth = np.linspace(min(r_fit_domain), max(r_fit_domain), 100)
            g_smooth = log_normal_shifted(r_smooth, *popt)
            
            plt.plot(r_smooth, g_smooth, 'r-', linewidth=2, label='Log-Normal Fit')
            
            # Calculate physical peak position from log-normal params
            physical_peak_r = np.exp(mu - sigma**2)
            
            plt.title(f"Peak Analysis (t={t_actual})\nPeak Location $\\approx$ {physical_peak_r:.2f}")
            print(f"Fit Params: Peak_Loc={physical_peak_r:.2f}, Width(sigma)={sigma:.3f}, Baseline={C:.3f}")
        
        plt.axhline(1.0, color='gray', linestyle=':')
        plt.legend()
        plt.show()
```
